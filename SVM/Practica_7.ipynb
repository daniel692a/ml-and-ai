```python
model = OneVsRestClassifier(SVC(kernel='linear', probability=True))
model.fit(X_train, y_train)

classes = model.classes_

for i, classifier in enumerate(model.estimators_):
    support_vectors = classifier.support_vectors_
    support_indices = classifier.support_
    y_train_binary = (y_train == classes[i]).astype(int)
    indices_positive = np.where(y_train_binary == 1)[0]
    indices_negative = np.where(y_train_binary == 0)[0]

    support_vectors_positive = support_vectors[indices_positive]
    support_vectors_negative = support_vectors[indices_negative]

    average_support_vector_positive = np.mean(support_vectors_positive, axis=0)
    average_support_vector_negative = np.mean(support_vectors_negative, axis=0)

    average_support_vector = (average_support_vector_positive + average_support_vector_negative) / 2
    norm_support = np.linalg.norm(average_support_vector)

    norm_support_positive = np.linalg.norm(average_support_vector_positive)
    norm_support_negative = np.linalg.norm(average_support_vector_negative)


    print(f"Clasificador para Clase {classes[i]} :")

    print(f'Numéro de elementos pertenecientes a la clase {classes[i]}: {len(support_indices)}')

    print(f"X_train:\n{X_train}")
    print(f"y_train (binario):\n{y_train_binary}\n")

    print(f"Vector promedio de soporte para {classes[i]}: {average_support_vector_positive}")
    print(f"Vector promedio de soporte para no {classes[i]}: {average_support_vector_negative}")
    print(f"c: {average_support_vector}")

    print(f"Norma del vector promedio de soporte para {classes[i]}: {norm_support_positive}")
    print(f"Norma del vector promedio de soporte para no {classes[i]}: {norm_support_negative}")
    print(f"Norma del vector promedio de soporte: {norm_support}\n")

    print("-" * 100)
```

```python
svm = SVC(kernel='linear', decision_function_shape='ovr', random_state=0, probability=True)
svm.fit(X_train, y_train)

def vector_norm(vector):
    return np.linalg.norm(vector)


for class_idx in range(len(iris.target_names)):
    support_vectors = svm.support_vectors_[svm.n_support_[:class_idx].sum():svm.n_support_[:class_idx + 1].sum()]

    non_support_vectors = np.delete(svm.support_vectors_, slice(svm.n_support_[:class_idx].sum(), svm.n_support_[:class_idx + 1].sum()), axis=0)

    mean_support_vector = np.mean(support_vectors, axis=0)
    norm_support_vector = vector_norm(mean_support_vector)

    mean_non_support_vector = np.mean(non_support_vectors, axis=0)
    norm_non_support_vector = vector_norm(mean_non_support_vector)

    hyperplane_vector = svm.coef_[class_idx]
    hyperplane_intercept = svm.intercept_[class_idx]

    print(f"Clase: {iris.target_names[class_idx]}")
    print(f"Número de instancias que pertenecen a la clase: {sum(y_train == class_idx)}")
    print(f"Vectores de soporte (clase): \n{support_vectors}")
    print(f"Vector promedio (clase): {mean_support_vector}")
    print(f"Norma del vector promedio (clase): {norm_support_vector}")
    print(f"Vectores de soporte (no clase): \n{non_support_vectors}")
    print(f"Vector promedio (no clase): {mean_non_support_vector}")
    print(f"Norma del vector promedio (no clase): {norm_non_support_vector}")
    print(f"Vector del hiperplano (coeficientes): {hyperplane_vector}")
    print(f"Intercepto del hiperplano: {hyperplane_intercept}")
    print("\n" + "-"*80 + "\n")
```

```python
y_pred = svm.predict(X_test)
y_prob = svm.predict_proba(X_test)

for i, (x, pred, prob) in enumerate(zip(X_test, y_pred, y_prob)):
    projections = [np.dot(x, svm.coef_[j]) + svm.intercept_[j] for j in range(len(iris.target_names))]
    print(f"Vector de prueba {i+1}: {x}")
    print(f"Proyección sobre el hiperplano por clase: {projections}")
    print(f"Probabilidades de pertenencia a cada clase: {prob}")
    print(f"Clasificación final: {iris.target_names[pred]}")
    print("\n" + "="*80 + "\n")
```
